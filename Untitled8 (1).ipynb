```python
import cv2
from google.colab.patches import cv2_imshow

def detect_moving_objects(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error: Couldn't open the video file.")
        return

    # Create background subtractor
    bg_subtractor = cv2.createBackgroundSubtractorMOG2()

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Apply the background subtractor and median blur
        fg_mask = cv2.medianBlur(bg_subtractor.apply(frame), 5)

        # Find contours in the mask
        contours, _ = cv2.findContours(fg_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for contour in contours:
            if cv2.contourArea(contour) > 100:
                # Calculate moments to find the center of the contour
                M = cv2.moments(contour)
                if M["m00"] != 0:
                    cx = int(M["m10"] / M["m00"])
                    cy = int(M["m01"] / M["m00"])

                    # Draw a dot (small filled circle) at the center of the contour
                    cv2.circle(frame, (cx, cy), 5, (255, 0, 0), -1)

        # Display the result
        cv2_imshow(frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(45) & 0xFF == ord('q'):
            break

    # Release the capture and close all OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

# Path to the video file
video_path = '/content/video_preview_h264.mp4'
detect_moving_objects(video_path)

```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Constants
MIN_CONTOUR_WIDTH = 40
MIN_CONTOUR_HEIGHT = 40
OFFSET = 10
LINE_HEIGHT = 550

# Variables
matches = []
vehicles = 0

def get_centroid(x, y, w, h):
    """Calculate the centroid of a bounding box."""
    x1 = int(w / 2)
    y1 = int(h / 2)
    cx = x + x1
    cy = y + y1
    return cx, cy

# Video capture setup
cap = cv2.VideoCapture('/content/istockphoto-981340772-640_adpp_is.mp4')
cap.set(3, 1920)  # Set width
cap.set(4, 1080)  # Set height

if not cap.isOpened():
    raise Exception("Failed to open video file.")

# Read the first two frames
ret, frame1 = cap.read()
if not ret:
    raise Exception("Failed to read the first frame.")
ret, frame2 = cap.read()

while ret:
    # Compute the difference between the two frames
    d = cv2.absdiff(frame1, frame2)
    grey = cv2.cvtColor(d, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(grey, (5, 5), 0)
    ret, th = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)
    dilated = cv2.dilate(th, np.ones((3, 3)))
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
    closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel)

    # Find contours
    contours, _ = cv2.findContours(closing, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    for i, c in enumerate(contours):
        (x, y, w, h) = cv2.boundingRect(c)
        contour_valid = (w >= MIN_CONTOUR_WIDTH) and (h >= MIN_CONTOUR_HEIGHT)

        if not contour_valid:
            continue

        # Draw bounding box and line
        cv2.rectangle(frame1, (x - 10, y - 10), (x + w + 10, y + h + 10), (255, 0, 0), 2)
        cv2.line(frame1, (0, LINE_HEIGHT), (1200, LINE_HEIGHT), (0, 255, 0), 2)

        centroid = get_centroid(x, y, w, h)
        matches.append(centroid)
        cv2.circle(frame1, centroid, 5, (0, 255, 0), -1)

        cx, cy = get_centroid(x, y, w, h)
        for (x, y) in matches:
            if LINE_HEIGHT - OFFSET < y < LINE_HEIGHT + OFFSET:
                vehicles += 1
                matches.remove((x, y))

    # Display vehicle count
    cv2.putText(frame1, f"Total Vehicle Detected: {len(contours)}", (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 170, 0), 2)

    # Show frame
    cv2_imshow(frame1)

    if cv2.waitKey(1) == 27:  # Exit on 'ESC' key
        break

    # Update frames
    frame1 = frame2
    ret, frame2 = cap.read()

# Cleanup
cv2.destroyAllWindows()
cap.release()
```

```python

import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow

# Function to calculate Euclidean distance between centers of two bounding boxes
def calculate_distance(bbox1, bbox2):
    # Calculate centers of the bounding boxes
    center1 = (bbox1[0] + bbox1[2] // 2, bbox1[1] + bbox1[3] // 2)
    center2 = (bbox2[0] + bbox2[2] // 2, bbox2[1] + bbox2[3] // 2)

    # Calculate Euclidean distance
    distance = np.sqrt((center1[0] - center2[0])**2 + (center1[1] - center2[1])**2)

    return distance

# Function to draw bounding box with color
def draw_bounding_box(image, bbox, color):
    cv2.rectangle(image, (bbox[0], bbox[1]), (bbox[0] + bbox[2], bbox[1] + bbox[3]), color, 2)

# Check if image file exists
image_path = '/content/social dist.jpg'

if not os.path.exists(image_path):
    print(f"Error: Image file '{image_path}' not found.")
else:
    # Load the image
    image = cv2.imread(image_path)

    # Check if image was successfully loaded
    if image is None:
        print(f"Error: Unable to load image '{image_path}'")
    else:
        # Define bounding boxes (x, y, width, height)
        bbox1 = (100, 50, 200, 150)
        bbox2 = (300, 200, 180, 120)

        # Draw individual bounding boxes
        draw_bounding_box(image, bbox1, (0, 255, 0))  # Green for bbox1
        draw_bounding_box(image, bbox2, (0, 255, 0))  # Green for bbox2

        # Calculate distance between bounding boxes
        distance = calculate_distance(bbox1, bbox2)

        # Determine color based on distance
        if distance < 200:  # Adjust threshold as needed
            color = (0, 255, 0)  # Green if distance is less
            print("Social Distancing")
        else:
            color = (0, 0, 255)  # Red if distance is more
            print("Not maintaining Social Distancing")

        # Draw bounding box around both objects based on distance
        bbox_combined = (min(bbox1[0], bbox2[0]), min(bbox1[1], bbox2[1]),
                         max(bbox1[0] + bbox1[2], bbox2[0] + bbox2[2]) - min(bbox1[0], bbox2[0]),
                         max(bbox1[1] + bbox1[3], bbox2[1] + bbox2[3]) - min(bbox1[1], bbox2[1]))

        draw_bounding_box(image, bbox_combined, color)

        # Display the image with bounding boxes and distance
        cv2.putText(image, f'Distance: {distance:.2f} pixels', (50, 30), cv2.FONT_HERSHEY_SIMPLEX, 1,
                    (255, 255, 255), 2)

        # Show the image in the notebook
        cv2_imshow(image)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
image = cv2.imread('/content/sun.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Perform Canny edge detection with adaptive thresholds
edges = cv2.Canny(blurred, threshold1=30, threshold2=100)

# Hough Circle Transform parameters
min_radius = 10
max_radius = 50

# Detect circles in the image
circles = cv2.HoughCircles(
    edges,
    cv2.HOUGH_GRADIENT,
    dp=1,
    minDist=20,
    param1=50,
    param2=30,
    minRadius=min_radius,
    maxRadius=max_radius
)

if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    for (x, y, r) in circles:
        # Draw the circle
        cv2.circle(image, (x, y), r, (0, 255, 0), 2)

# Display results
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("/lane.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
gray_img = cv2.dilate(gray_img, kernel=np.ones((5, 5), np.uint8))
canny = cv2.Canny(gray_img, 100, 200)

roi_vertices = [(270, 670), (600, 400), (1127, 712)]


def roi(image, vertices):
  mask = np.zeros_like(image)
  mask_color = 255
  cv2.fillPoly(mask, vertices, mask_color)
  masked_img = cv2.bitwise_and(image, mask)
  return masked_img

roi_image = roi(canny, np.array([roi_vertices], np.int32))


lines = cv2.HoughLinesP(roi_image, 1, np.pi/180, 100, minLineLength=100, maxLineGap=10)


def draw_lines(image, hough_lines):
  for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
  return image

final_img = draw_lines(img, lines)
plt.imshow(final_img)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# Install required modules beforehand:
# !pip install opencv-contrib-python
# !pip install opencv-python
# !pip install numpy
# !pip install times

import cv2
import numpy as np
from time import sleep
from google.colab.patches import cv2_imshow

# Define parameters
largura_min = 80  # Minimum width of the rectangle
altura_min = 80   # Minimum height of the rectangle
offset = 6        # Allowed error between pixels
pos_linha = 550   # Position of the counting line
delay = 60        # FPS of the video
detec = []
carros = 0

# Function to get the center of a bounding box
def pega_centro(x, y, w, h):
    x1 = int(w / 2)
    y1 = int(h / 2)
    cx = x + x1
    cy = y + y1
    return cx, cy

# Open the video file
cap = cv2.VideoCapture('/content/istockphoto-981340772-640_adpp_is.mp4')

# Create background subtractor
subtracao = cv2.bgsegm.createBackgroundSubtractorMOG()

while True:
    ret, frame1 = cap.read()
    if not ret:
        break

    # Calculate sleep time to match FPS
    tempo = float(1 / delay)
    sleep(tempo)

    # Process the frame
    grey = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(grey, (3, 3), 5)
    img_sub = subtracao.apply(blur)

    # Morphological operations
    dilat = cv2.dilate(img_sub, np.ones((5, 5)))
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    dilatada = cv2.morphologyEx(dilat, cv2.MORPH_CLOSE, kernel)
    dilatada = cv2.morphologyEx(dilatada, cv2.MORPH_CLOSE, kernel)

    # Find contours
    contorno, h = cv2.findContours(dilatada, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Draw counting line
    cv2.line(frame1, (25, pos_linha), (1200, pos_linha), (255, 127, 0), 3)

    # Process each contour
    for i, c in enumerate(contorno):
        x, y, w, h = cv2.boundingRect(c)
        validar_contorno = (w >= largura_min) and (h >= altura_min)
        if not validar_contorno:
            continue

        # Draw bounding box and center
        cv2.rectangle(frame1, (x, y), (x + w, y + h), (0, 255, 0), 2)
        centro = pega_centro(x, y, w, h)
        detec.append(centro)
        cv2.circle(frame1, centro, 4, (0, 0, 255), -1)

    # Count vehicles
    for (x, y) in detec:
        if y < (pos_linha + offset) and y > (pos_linha - offset):
            carros += 1
            detec.remove((x, y))

    # Draw updated counting line
    cv2.line(frame1, (25, pos_linha), (1200, pos_linha), (0, 127, 255), 3)

    # Display the count
    print("Car is detected: ")
    print(len(contorno))
    cv2.putText(frame1, "VEHICLE COUNT: " + len(contorno), (450, 70), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 5)

    # Show frames
    cv2_imshow(frame1)
    cv2_imshow(dilatada)

    # Break on 'ESC' key press
    if cv2.waitKey(1) == 27:
        break

# Release resources
cv2.destroyAllWindows()
cap.release()
```

```python
import qrcode
import cv2
from PIL import Image
import numpy as np
import os

QR_FOLDER = 'generated_qr_codes/'

# Create QR_FOLDER if it does not exist
if not os.path.exists(QR_FOLDER):
    os.makedirs(QR_FOLDER)

def generate_qr(data, filename):
    qr = qrcode.QRCode(version=1, box_size=10, border=5)
    qr.add_data(data)
    qr.make(fit=True)
    img = qr.make_image(fill='black', back_color='white')
    img_filename = os.path.join(QR_FOLDER, filename)
    img.save(img_filename)
    img.show()
    print(f"QR code generated and saved as {img_filename}")

def decode_qr(filepath):
    # Load the image
    img = cv2.imread(filepath)

    # Create a QRCode detector object
    detector = cv2.QRCodeDetector()

    # Detect and decode the QR code
    retval, decoded_info, points, _ = detector(img)

    if retval:
        print(f"Decoded data from QR code: {decoded_info}")
        return decoded_info
    else:
        print("No valid QR code found.")
        return []

def main():
    while True:
        print("\n1. Generate QR Code")
        print("2. Decode QR Code")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            data = input("Enter the data to encode in the QR code: ")
            filename = input("Enter the filename to save the QR code image (e.g., qr_code.png): ")
            generate_qr(data, filename)
        elif choice == '2':
            filepath = input("Enter the filepath of the QR code image to decode: ")
            if os.path.exists(filepath):
                decode_qr(filepath)
            else:
                print("File does not exist.")
        elif choice == '3':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == '__main__':
    main()
```
